## Import Libraries

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("kc_house_data.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe().T
```

```python
plt.figure(figsize=(18,10))
sns.heatmap(data.corr(),annot=True,cmap="mako")
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=data,x=data["view"],y=data["price"])
```

```python
plt.figure(figsize=(8,4))
sns.lineplot(data=data,x=data["view"],y=data["price"])
```

```python
plt.figure(figsize=(7,3))
sns.boxplot(x="waterfront", y="price", data=data)
```

```python
sns.scatterplot(data=data,x=data["bedrooms"],y=data["price"])
```

```python
sns.scatterplot(data=data, x=data["bathrooms"], y=data["price"])
```

```python
sns.regplot(x="price", y="sqft_living", data=data)
plt.ylim(0,)
```

```python
sns.regplot(x="price", y="sqft_lot", data=data)
plt.ylim(0,)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=data["floors"],y=data["price"])
```

```python
plt.figure(figsize=(10,5))
sns.countplot(y = data["condition"])
```
